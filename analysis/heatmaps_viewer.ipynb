```python
data = {
    2023: {"cluster_file": "clusters_2023_sustantivas_5_v3", "tree_file": "../data/estructura/2023_12_09.csv"},
    2025: {"cluster_file": "clusters_2025_sustantivas_5_v3", "tree_file": "../data/estructura/2025_07_08.csv"},
}
```

```python
year = 2023
cluster_id = 6
central_administration_only = True
threshold = 0.7
dimensions_weights={
    "objetivos": 0.5,
    "distancia": 0.4,
    "destinatarios": 0.08,
    "ambitos": 0.02,
}
```

```python
from chainsaw.heatmaps.plot import heatmap

heatmap(
    clusters_file_name=data[year]["cluster_file"],
    cluster_id=cluster_id,
    tree_file=data[year]["tree_file"],
    central_administration_only=central_administration_only,
    ellipsis=50,
    threshold=threshold,
    dimensions_weights=dimensions_weights,
    width=900,
    height=900,
    full=False,
    strict=True,
)
```
